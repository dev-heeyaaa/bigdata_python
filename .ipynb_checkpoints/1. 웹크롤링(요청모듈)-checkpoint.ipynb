```python
url = 'https://www.naver.com'

res = rq.get(url)

# 응답 코드 가져오기
print(res.status_code)
```

```python
# 응답코드를 활용한 분기

def url_check(url):
    res = rq.get(url)
    
    print(res)
    
    sc = res.status_code
    
    if sc == 200:
        print(f'{url} 요청 성공')
        
    elif sc == 404:
        print(f'{url} 찾을 수 없음')
        
    else:
        print(f'{url} 알 수 없는 에러 : {sc}')
```

```python
url_check('https://www.naver.com')
```

## urllib으로 웹페이지 접속

```python
from urllib.request import urlopen
```

```python
url = 'https://www.naver.com'

page = urlopen(url) # 요청 객체를 이용해 요청

print(page)
```

```python
print(page.code)
print(page.headers)
print(page.url)
print(page.info().get_content_charset())
```

```python
print(page.read())
```

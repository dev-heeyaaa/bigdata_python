## inner join 과 outer join
- 내부 조인 (inner join)
    - 둘 이상의 데이터프레임에서 조건에 맞는 행을 연결하는 것
  
- 외부 조인 (outer join)
     - 데이터 프레임 중 어떤 프레임을 기준으로 할 것인지에 따라 왼쪽 외부 조인(left outer join), 오른쪽 외부 조인(right outer join), 완전외부 조인(Full outer jon)으로 나뉨
         - 왼쪽 외부 조인 : 왼쪽 데이터 프레임을 모두 포함하여 연결
         - 오른쪽 외부 조인 : 오른쪽 데이터프레임을 모두 포함하여 연결
         - 완전 외부 조인 : 왼쪽과 오른쪽 데이터 프레임을 모두 포함하여 연결

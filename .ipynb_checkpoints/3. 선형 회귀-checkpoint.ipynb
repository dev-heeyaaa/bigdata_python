```python
import pandas as pd
import numpy as np # 행렬 연산
import matplotlib.pyplot as plt #시각화 라이브러리
import seaborn as sns # 시각화 라이브러리
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # 선형 모델 선형 회귀
from sklearn.metrics import mean_squared_error # 모델 평가하는
```

# 데이터 준비

```python
# csv 파일을 데이터프레임으로 변환

df = pd.read_csv('./data/auto-mpg.csv', header = None)
```

```python
df.head()
```

```python
# 열 이름 지정
df.columns = ['mpg', 'cylinders','displacement','horsepower','weight','acceleration','model_year','origin','name']
```

```python
# 데이터 자료형 확인
df.dtypes
```

```python
# 데이터 통계 요약 정보 확인
df.describe()
```

```python
# horsepower 열의 자료형 변경 (문자열 -> 숫자)

df['horsepower'].unique() # horsepower 열의 고유값 확인
```

## 변수 선택

```python
# 분석에 활용할 열 선택(연비, 실린더, 출력, 중량)
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]
ndf.head()
```

```python
# 종속 변수 Y (mpg)와 다른 변수 간의 선형관계를 그래프로 확인
# matplotlib으로 산점도 그리기
ndf.plot(kind = 'scatter', x = 'weight', y = 'mpg', c = 'coral', s = 10, figsize = (10, 5))
plt.show()
plt.close()
```

```python
# seaborn 으로 산점도 그리기
fig = plt.figure(figsize = (10, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
sns.regplot(x = 'weight', y = 'mpg', data = ndf, ax = ax1) # 회귀선 표시
sns.regplot(x = 'weight', y = 'mpg', data = ndf, ax = ax2, fit_reg = False) # 회귀선 미표시

plt.show()
plt.close()
```

```python
# seaborn 조인트 그래프 - 산점도, 히스토그램

sns.jointplot(x = 'weight', y = 'mpg', data = ndf)
sns.jointplot(x = 'weight', y = 'mpg', data = ndf, kind = 'reg') # 회귀선 표시

plt.show()
plt.close()
```

```python
# seaborn pairplot으로 두 변수 간의 모든 경우의 수 그리기
sns.pairplot(ndf)

# 장점은 한 번에 다 볼 수 있다
# 단점 변수가 많아지면 많아질 수록 연산 처리량이 많아져서 연산 속도가 느리다

plt.show()
plt.close()
```

# 데이터셋 구분 - train data/ test data

```python
# 변수 선택
# 독립 변수 x1         x2            x3
x = ndf[['weight', 'horsepower', 'cylinders']]
y = ndf[['mpg']] # 종속 변수 y
```

```python
# train data와 test data로 구분 (7:3)          
x_train, x_test, y_train, y_test = train_test_split(x, # 독립변수
                                                    y, # 종속변수
                                                    test_size = 0.3, # 테스트 30%
                                                    random_state = 12) # 랜덤 추출 값. 난수값 고정하기
```

```python
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

# 회귀 분석 모형 - sklearn 사용

```python
# 단순 회귀 분석 모형 객체 생성
lr = LinearRegression()
```

```python
# train data 를 가지고 모형 학습
lr.fit(x_train, y_train)
```

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수 계산
r_square = lr.score(x_test, y_test)
r_square
```

```python
# 회귀식의 기울기
lr.coef_
```

```python
# 회귀식의 y 절편
lr.intercept_
```

```python
# Lr 모형에 x_test를 입력하여 예측한 값을 실제 값과 비교
y_pred = lr.predict(x_test) # predict - 추론하다 (모델 추론)
rmse = mean_squared_error(y_test, y_pred, squared = False)
rmse
```

```python
pd.DataFrame({'ans':y_test['mpg'], 'pred':y_pred.flatten()})
```
